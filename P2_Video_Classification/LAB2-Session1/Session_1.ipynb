Debido a la componente de aletoriedad del modelo aleatoria, definimos una función sencilla para poder hacer varias ejecuciones del clasisficador y obtener una media de su accuracy.

```
import numpy as np
def execute_radom_vs_fixed(folder_data='data',n_executions=1000, fix_mode = 'ApplyEyeMakeup' ):
  """Executes n_executions apply_random_vs_fixed_mode from random_vs_fixed_mode.py
  """
  radom_accuracies = []
  for i in range(n_executions):
    random_accuracy, mode_accuracy = apply_random_vs_fixed_mode(folder_data,fix_mode,verbosse = False )
    radom_accuracies.append(random_accuracy)

  r_acc_mean = np.mean(radom_accuracies) *100
  r_acc_std = np.std(radom_accuracies) 
  fixed_acc = mode_accuracy *100
  print('Random Accuracy Mean', "{:.2f}".format(r_acc_mean),  '%')
  print('Random Accuracy Std',"{:.2f}".format(r_acc_std))
  print('Mode matched:',"{:.2f}".format(fixed_acc) ,  '%')
```
